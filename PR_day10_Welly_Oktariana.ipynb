Handling outlier

```python
from scipy import stats
```

```python
def check_plot(df, variable):
    # fungsi mengambil kerangka data (df) dan
    # variabel yang diminati sebagai argumen

    # tentukan ukuran gambar
    plt.figure(figsize=(16, 4))

    # histogram
    plt.subplot(1, 3, 1)
    sns.histplot(df[variable], bins=30)
    plt.title('Histogram')

    # plot Q-Q
    plt.subplot(1, 3, 2)
    stats.probplot(df[variable], dist="norm", plot=plt)
    plt.ylabel('Variable quantiles')

    # box plot
    plt.subplot(1, 3, 3)
    sns.boxplot(y=df[variable])
    plt.title('Boxplot')

    plt.show()
```

```python
#outlier visualization
column = train.drop(columns ='Concrete compressive strength(MPa, megapascals) ').columns

for col in column:
  check_plot(train,col)
```

```python
#handling outlier
def find_iqr(df, variable):
    IQR = df[variable].quantile(0.75) - df[variable].quantile(0.25)
    lower_boundary = df[variable].quantile(0.25) - (IQR * 1.5)
    upper_boundary = df[variable].quantile(0.75) + (IQR * 1.5)
    q25= df[variable].quantile(0.25)
    q75 = df[variable].quantile(0.75)

    return lower_boundary, upper_boundary,IQR, q25, q75

for col in column:
  data_atas, data_bawah,IQR, q25, q75= find_iqr(train,col)
  train[col]= np.where(train[col]<=q25,data_bawah,
                              np.where(train[col]>=q75,data_atas,train[col]))
  test[col]= np.where(test[col]<=q25,data_bawah,
                              np.where(test[col]>=q75,data_atas,test[col]))
```

```python
#visualization data after handling outlier
for col in column:
  check_plot(train,col)
```

# Modeling

```python
import sklearn as sk
from sklearn.linear_model import Lasso, LinearRegression, Ridge
from sklearn.metrics import mean_squared_error, r2_score
import numpy
```

```python
X_train = train.drop(columns=['Concrete compressive strength(MPa, megapascals) '])
y_train = train['Concrete compressive strength(MPa, megapascals) ']
X_test = test.drop(columns=['Concrete compressive strength(MPa, megapascals) '])
y_test = test['Concrete compressive strength(MPa, megapascals) ']
```

```python
lasso_model = Lasso() #Lasso
lasso_model.fit(X_train, y_train)
```

```python
#prediksi memakai lasso di train
y_train_pred_lasso = lasso_model.predict(X_train)

#RMSE Lasso
rmse_train_lasso = np.sqrt(mean_squared_error(y_train, y_train_pred_lasso))

#Prediksi memakai lasso di test
y_pred_lasso = lasso_model.predict(X_test)
rmse_test_lasso = np.sqrt(mean_squared_error(y_test, y_pred_lasso))

# R2 Score
r2_score_train = r2_score(y_train,y_train_pred_lasso)
r2_score_test = r2_score(y_test,y_pred_lasso)

print("RMSE on Training Data:", rmse_train_lasso)
print("RMSE on Testing Data :", rmse_test_lasso)
```

Hasil RMSE tidak lebih dar 10% antara training dan testing, sehingga modle cukup baik. Namun kit acoba improve model

```python
from sklearn.model_selection import GridSearchCV

# Parameter yang akan diuji, bisa divariasikan sesuai dengan dokumentasi
param_grid = {
    'alpha': [0.0001,0.1, 1.0, 10.0]}


# Inisialisasi model Lasso
Lasso_model = Lasso()

# Inisialisasi objek GridSearchCV
grid_search = GridSearchCV(Lasso_model, param_grid, cv=5, scoring='neg_mean_squared_error')

# Lakukan pencarian parameter terbaik
grid_search.fit(X_train, y_train)
```

```python
# Tampilkan parameter terbaik
best_params = grid_search.best_params_
print("Parameter Terbaik:", best_params)
```

```python
best_params = grid_search.best_params_
lasso_grid_best = Lasso(**best_params)
lasso_grid_best.fit(X_train, y_train)
```

```python
# Predict the target values for the training and testing data
y_train_lasso_pred = lasso_grid_best.predict(X_train)
y_test_lasso_pred = lasso_grid_best.predict(X_test)

# Calculate RMSE for training and testing data
rmse_train = np.sqrt(mean_squared_error(y_train, y_train_lasso_pred))
rmse_test = np.sqrt(mean_squared_error(y_test, y_test_lasso_pred))


print("RMSE on training set:", rmse_train)
print("RMSE on testing set:", rmse_test)
```

hasil tuning model tidak terlalu significan, artinya performa model sudah mencapai akurasi yang mendekati maksimal
